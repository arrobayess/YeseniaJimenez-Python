```python
import modulo1, modulo2, modulo3, modulo4, modulo5, modulo6, modulo7

print("Elije un módulo:")
print("1. Módulo 1 suma")
print("2. Módulo 2 multiplicación")
print("3. Módulo 3 potencia")
print("4. Módulo 4 area de rectangulo")
print("5. Módulo 5 resta")
print("6. Módulo 6 area triangulo")
print("7. Módulo 7 división")

opcion = int(input("Ingresa el número del módulo: "))

def obtener_datos_entrada():
    if opcion == 1:
        a = float(input("Ingresa el primer número: "))
        b = float(input("Ingresa el segundo número: "))
        return a, b
    elif opcion == 2:
        a = float(input("Ingresa el primer número: "))
        b = float(input("Ingresa el segundo número: "))
        return a, b
    elif opcion == 3:
        base = float(input("Ingresa la base: "))
        exponente = float(input("Ingresa el exponente: "))
        return base, exponente
    elif opcion == 4:
        largo = float(input("Ingresa el largo del rectángulo: "))
        ancho = float(input("Ingresa el ancho del rectángulo: "))
        return largo, ancho
    elif opcion == 5:
        a = float(input("Ingresa el primer número: "))
        b = float(input("Ingresa el segundo número: "))
        return a, b
    elif opcion == 6:
        base = float(input("Ingresa la base del triángulo: "))
        altura = float(input("Ingresa la altura del triángulo: "))
        return base, altura
    elif opcion == 7:
        a = float(input("Ingresa el primer número: "))
        b = float(input("Ingresa el segundo número (divisor): "))
        return a, b
    else:
        print("Opción no válida")
        return None

if opcion in range(1, 8):
    if opcion == 1:
        datos = obtener_datos_entrada()
        resultado = modulo1.suma(*datos)
    elif opcion == 2:
        datos = obtener_datos_entrada()
        resultado = modulo2.multiplicacion(*datos)
    elif opcion == 3:
        resultado = modulo3.potencia(*datos)
    elif opcion == 4:
        datos = obtener_datos_entrada()
        resultado = modulo4.area_rectangulo(*datos)
    elif opcion == 5:
        datos = obtener_datos_entrada()
        resultado = modulo5.resta(*datos)
    elif opcion == 6:
        datos = obtener_datos_entrada()
        resultado = modulo6.area_triangulo(*datos)
    elif opcion == 7:
        datos = obtener_datos_entrada()
        resultado = modulo7.division(*datos)

    if datos:
        print("Resultado:", resultado)
```
